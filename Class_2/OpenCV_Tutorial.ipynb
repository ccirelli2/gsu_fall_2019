# Example Problem
    - Extract the two pictures and determine some properties of the images

```python
image = cv2.imread('wall_picture.JPG')
imshow(image)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
imshow(hsv) # says probably can't use HSV
```

```python
blurred_image = cv2.GaussianBlur(image, (3,3), 1)
imshow(blurred_image)
```

```python
canny_edge = cv2.Canny(blurred_image, 50, 5)
imshow(canny_edge)
```

```python
# Get Mask / Expand edges
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (100,100))  
closed_img = cv2.morphologyEx(canny_edge, cv2.MORPH_CLOSE, kernel)
imshow(closed_img)
```

```python
# Now find contours
contours = cv2.findContours(closed_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) # find external contours
# use simple contours to only return the basic contours. 
contours
```

```python
# Sort Contours
sorted_contours = sorted(contours, key=cv2.contourArea, reverse=True)[:2] # get last two
```

```python
# now draw contours, use cv2.drawContours(contour_output, sorted_contours, -1, 255, -1)
```
